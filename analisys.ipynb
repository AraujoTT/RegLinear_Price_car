```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('data/clean_data.csv')
df.head()
```

```python
df.describe().round(2)
```

```python
df.corr().round(2)
```

```python
ax = sns.boxplot(data=df['Price'], orient='v', width=0.3)
ax.figure.set_size_inches(12, 6)
ax.set_title('Vehicle price', fontsize = 16)
ax.set_ylabel('Price ($)',fontsize= 12)
```

```python
ax = sns.boxplot(data=df, y='Price', x='Engine_cyl_num', orient='v' )
ax.figure.set_size_inches(12, 6 )
```

```python
ax = sns.boxplot(data=df, y= 'Price', x= 'Year', orient='v')
ax.figure.set_size_inches(12, 6)
```

```python
ax = sns.histplot(df['Price'], kde=True)
ax.figure.set_size_inches(12, 6)
```

```python
ax = sns.pairplot(df, y_vars='Price', x_vars=['Year', 'FuelConsumption(L/KM)', 'Kilometres', 'Doors', 'Seats', 'Automatic', 'Manual', 'DEMO', 'NEW', 'USED', 'Engine_cyl_num',	'Engine_total_vol'], height= 5, kind= 'reg')
ax.fig.suptitle('dispersion between variables', fontsize=16, y=1.05)
```

```python
df['log_Price'] = np.log(df['Price'])
df['log_Year'] = np.log(df['Year'])
df['log_FuelConsumption(L/KM)'] = np.log(df['FuelConsumption(L/KM)'])
df['log_Kilometres'] = np.log(df['Kilometres'])
df['log_Doors'] = np.log(df['Doors'])
df['log_Seats'] = np.log(df['Seats'])
df['log_Automatic'] = np.log(df['Automatic'] +1)
df['log_Manual'] = np.log(df['Manual'] +1)
df['log_DEMO'] = np.log(df['DEMO'] +1)
df['log_NEW'] = np.log(df['NEW'] +1)
df['log_USED'] = np.log(df['USED'] +1)
df['log_Engine_cyl_num'] = np.log(df['Engine_cyl_num'])
df['log_Engine_total_vol'] = np.log(df['Engine_total_vol'])
```

```python
ax = sns.pairplot(df, y_vars='log_Price', x_vars=['log_Year', 'log_FuelConsumption(L/KM)', 'log_Kilometres', 'log_Doors', 'log_Seats', 'log_Automatic', 'log_Manual', 'log_DEMO', 'log_NEW', 'log_USED', 'log_Engine_cyl_num',	'Engine_total_vol'], height= 5, kind= 'reg')
ax.fig.suptitle('dispersion between variables', fontsize=16, y=1.05)
ax
```

```python
ax = sns.histplot(df['log_Price'], kde=True)
ax.figure.set_size_inches(12, 6)
```

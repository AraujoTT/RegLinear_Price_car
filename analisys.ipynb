```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('data/clean_capping_data.csv')
df.head()
```

```python
df.describe().round(2)
```

```python
df['Price'] = np.log(df['Price'])
df.corr().round(2)
```

```python
ax = sns.boxplot(data=df['Price'], orient='v', width=0.3)
ax.figure.set_size_inches(12, 6)
ax.set_title('Vehicle price', fontsize = 16)
ax.set_ylabel('Price ($)',fontsize= 12)
```

```python
ax = sns.boxplot(data=df, y='Price', x='Engine_cyl_num', orient='v' )
ax.figure.set_size_inches(12, 6 )
```

```python
ax = sns.boxplot(data=df, y= 'Price', x= 'Year', orient='v')
ax.figure.set_size_inches(12, 6)
```

```python
ax = sns.histplot(df['Price'], kde=True)
ax.figure.set_size_inches(12, 6)
```

```python
ax = sns.pairplot(df, y_vars='Price', x_vars=['Year', 'FuelConsumption', 'Kilometres', 'Doors', 'Seats', 'Automatic', 'Manual', 'DEMO', 'NEW', 'USED', 'Engine_cyl_num',	'Engine_total_vol'], height= 5, kind= 'reg')
ax.fig.suptitle('dispersion between variables', fontsize=16, y=1.05)
```

```python
df.to_csv('data/clean_log_capping_data.csv', index=False)
```
